## Padrão de card testing

Clientes 2007, 2024 e 2038 fizeram microtransações (R$2 a R$4) logo antes de uma compra grande.
Esse padrão é chamado de card testing: o fraudador valida o cartão com valores pequenos antes de usar de verdade.

## Regra de detecção

Três fatores se destacaram na análise:
- valor acima de R$5.000
- transação antes das 5h
- localização internacional presente no dataset

```python
with open('fraude.py', 'w') as f:
    f.write("""
LOCAIS_SUSPEITOS = ['Moscou', 'Toquio', 'Lagos', 'Amsterda', 'Berlim']

def detectar_fraude_regra(row):
    if row['valor'] > 5000:
        return 1
    if row['hora'] < 5:
        return 1
    if row['localizacao'] in LOCAIS_SUSPEITOS:
        return 1
    return 0
""")
```

Acurácia das regras no dataset:

```python
LOCAIS_SUSPEITOS = ['Moscou', 'Toquio', 'Lagos', 'Amsterda', 'Berlim']

df['fraude_prevista'] = df.apply(
    lambda row: 1 if row['valor'] > 5000 or row['hora'] < 5
    or row['localizacao'] in LOCAIS_SUSPEITOS else 0, axis=1
)

acuracia = (df['fraude_prevista'] == df['fraude']).mean()
print(f'Acuracia: {acuracia:.0%}')
```
